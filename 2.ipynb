```python
# -*- coding: utf-8 -*-
"""

Get some chips

"""
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd
dataBase = pd.read_csv("pirate_attacks.csv")
filePath = r"pirate_attacks.csv"
dataBase = pd.read_csv(filePath)

fig = plt.figure(figsize=(10, 8), dpi=100)
ax = plt.axes(projection=ccrs.PlateCarree())

ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.LAKES, edgecolor='royalblue')
ax.add_feature(cfeature.RIVERS)
ax.stock_img()

lon = dataBase['longitude']
lat = dataBase['latitude']
ax.scatter(lon, lat, transform=ccrs.PlateCarree(), color='red', s=1)


fig.savefig(r'.png')

plt.show()
```
