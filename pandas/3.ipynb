```python
# -*- coding: utf-8 -*-
"""

Get some chips

"""

import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
plt.rc('font', family='Times New Roman', size=20)
filePath = r"pirate_attacks.csv"
dataBase = pd.read_csv(filePath)

date = [I[:4] for I in dataBase['date']]
dataBase['date'] = date

dateCount = Counter(date)

plt.figure(figsize=(10, 8), dpi=100)
plt.bar(dateCount.keys(), dateCount.values(), color='black')
plt.xlabel('Year')
plt.ylabel('Number')
plt.xticks(rotation=90)
plt.savefig(r'柱状图.png')
plt.show()
plt.clf()
```
